```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt

# Load 
coffee_near_ev_wa_df = pd.read_csv("resources/coffee_info_tesla_ev.csv")

# Display
coffee_near_ev_wa_df.head()
```

```python
print(coffee_near_ev_wa_df.head())
```

```python
# Count the occurrences of each restaurant name
restaurant_counts = coffee_near_ev_wa_df['Restaurant Name'].value_counts()

# Display
print(restaurant_counts)
```

```python
# Get the top 5 restaurants
top_5_restaurants = restaurant_counts.head(5)

# Display
print(top_5_restaurants)
```

```python
# Data
restaurant_names = top_5_restaurants.index
frequency = top_5_restaurants.values

# Plot
plt.figure(figsize=(10, 6))
plt.bar(restaurant_names, frequency, color='skyblue')
plt.title('Top 5 Coffee Shops Near EV Stations in Washington')
plt.xlabel('Coffee Shop')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count occurrences of each restaurant
restaurant_counts = coffee_near_ev_wa_df['Restaurant Name'].value_counts()

# Get the count of Starbucks
starbucks_count = restaurant_counts.get('Starbucks', 0)

# Get the count of everything else
other_count = restaurant_counts.sum() - starbucks_count

print("Starbucks Count:", starbucks_count)
print("Other Count:", other_count)
```

```python
# Data
coffee_shop_counts = {'Starbucks': 116, 'Dutch Bros. Coffee': 11, 'Other': 203}

# Plot
plt.figure(figsize=(8, 8))
plt.pie(coffee_shop_counts.values(), labels=coffee_shop_counts.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Coffee Shop Distribution in WA')
plt.axis('equal')

# Save the plot as a PNG file
plt.savefig('resources/chart_coffee_distribution_wa.png')

# Show
plt.show()
```
